```python
import cv2
import os
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
import numpy as np
import warnings
    
# 忽略警告
warnings.simplefilter("ignore")
warnings.filterwarnings("ignore", message="Unable to register cuDNN factory", category=UserWarning)
warnings.filterwarnings("ignore", message="Unable to register cuFFT factory", category=UserWarning)
warnings.filterwarnings("ignore", message="Unable to register cuBLAS factory", category=UserWarning)

# 初始化圖像和標籤列表
images = []
labels = []

print('開始讀檔')

#讀取檔案內容
for filename in os.listdir('/kaggle/input/ml-hw2/HW2_MNIST_train'):
    #讀取文字檔案
    if filename.endswith('.txt'):
        labels_file_path = os.path.join('/kaggle/input/ml-hw2/HW2_MNIST_train', filename)
        with open(labels_file_path, 'r') as file:
            lines = file.readlines()
            #print(lines)
            
        # 處理每張圖像
        for line in lines:
            
            #用\t切出每個元素
            parts = line.strip().split('\t')
            
            #依序是圖片數字、左上角x跟y座標、物件寬度及高度
            digit = int(parts[0])
            pos_x = float(parts[1])
            pos_y = float(parts[2])
            img_w = float(parts[3])
            img_h = float(parts[4])
            
            #讀取圖片
            image_path = os.path.join('/kaggle/input/ml-hw2/HW2_MNIST_train', filename.replace('.txt', '.png'))
            
            #讀取灰階圖檔，做resize、擴展維度和歸一化
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (64, 64))
            if img is None:
                print(f"Error reading image: {image_path}")
                continue
            digit_img = img.astype(np.float32) / 255.0
            images.append(digit_img)
            labels.append(digit)
            break

print('讀檔完畢，開始訓練')
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Input, Dense, SimpleRNN, LSTM, GRU, Flatten
from keras.models import Sequential
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.models import load_model

def show_train_history(train_history, train, validation):  
    plt.plot(train_history.history[train])  
    plt.plot(train_history.history[validation])  
    plt.title('Train History')  
    plt.ylabel(train)  
    plt.xlabel('Epoch')  
    plt.legend(['train', 'validation'], loc='upper left')  
    plt.show()
# Function definitions for plotting
def show_train_history_all(train_histories, metrics, title):
    plt.figure(figsize=(10, 5))
    for metric, histories in train_histories.items():
        for history, color in zip(histories, colors):
            plt.plot(history.history[metric], color=color, label=history.name)
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel(metrics)
    plt.legend()
    plt.grid(True)
    plt.show()

# 轉換成numpy型態
images_np = np.array(images)
labels_np = np.array(labels)

# one-hot編碼
labels_one_hot = to_categorical(labels_np, num_classes=10)

# 切割數據集
train_images, val_images, train_labels, val_labels = train_test_split(images_np, labels_one_hot, test_size=0.2, random_state=42)


# Define model paths
model_paths = ['saved_model_simpleRnn.h5', 'saved_model_lstm.h5', 'saved_model_gru.h5']

# Initialize accumulated validation data
val_images_accumulated = []
val_labels_accumulated = []

# Loop through models
models = []
train_histories = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': []}
colors = ['blue', 'orange', 'green']

for i, (model_type, model_path) in enumerate(zip([SimpleRNN, LSTM, GRU], model_paths)):
    model_name = model_type.__name__
    
    # Check if model file exists and load it
    if os.path.exists(model_path):
        model = load_model(model_path)
        print(f"Loaded model '{model_name}' from '{model_path}'.")
    else:
        print(f"Model file '{model_path}' not found. Training {model_name} model...")
        # Build and compile the model
        model = Sequential()
        model.add(model_type(units=256, input_shape=(64,64), activation='relu', return_sequences=True))
        model.add(Flatten())
        model.add(Dense(10, activation='softmax'))
        model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
        
        # Train the model
        train_history = model.fit(train_images, train_labels, epochs=10, batch_size=128, validation_data=(val_images, val_labels))
        train_history.name = model_name
        train_histories['loss'].append(train_history)
        train_histories['val_loss'].append(train_history)
        train_histories['accuracy'].append(train_history)
        train_histories['val_accuracy'].append(train_history)
        
        # Save the model
        model.save(model_path)
        print(f"Model '{model_name}' trained and saved to '{model_path}'.")
    
    # Evaluate the model
    loss, accuracy = model.evaluate(val_images, val_labels, verbose=0)
    print(f"{model_name} - Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
    
    # Store the model
    models.append(model)
    # Plotting
    show_train_history(train_history, 'accuracy', 'val_accuracy')
    show_train_history(train_history, 'loss', 'val_loss')


# Plot training history all
show_train_history_all(train_histories, 'Loss and Accuracy', 'Training History Comparison')
```

## **根據上述三種模型accuracy和loss的比較結果，我們最後是選擇GRU的模型**

```python
import os
import cv2
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model

# 刪除檔案
model_path = 'HW4_prob1.csv'
if os.path.exists(model_path):
    os.remove(model_path)
    print(f"檔案 '{model_path}' 已成功刪除。")
else:
    print(f"檔案 '{model_path}' 不存在，無需刪除。")

# 使用GRU模型訓練
model = load_model('/kaggle/working/saved_model_gru.h5')

test_folder_path = '/kaggle/input/ml-hw2/HW2_MNIST_test'

# 初始化
test_image_paths = []
predictions = []

for filename in os.listdir(test_folder_path):
    # 讀取圖片
    if filename.endswith('.png'):
        image_path = os.path.join(test_folder_path, filename)
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (64, 64))
        if img is None:
            print(f"Error reading image: {image_path}")
            continue
        digit_img = img.astype(np.float32) / 255.0

        # 預測label
        pred = model.predict(np.expand_dims(digit_img, axis=0), verbose=0)
        predicted_label = np.argmax(pred)

        # 儲存圖片跟預測結果
        test_image_paths.append(filename)
        predictions.append(predicted_label)

# 建立dataframe
results_df = pd.DataFrame({'image': test_image_paths, 'class': predictions})

# 依照class去排序
results_df = results_df.sort_values(by='class')

results_df.to_csv('HW4_prob1.csv', index=False)

print('完成預測檔 HW4_prob1.csv')
```
